```python
plt.figure(figsize=(10,6))
colors = plt.cm.viridis(np.linspace(0, 1, len(top10_states)))
plt.bar(top10_states.index, top10_states.values,width=0.7,color=colors)
 
plt.xlabel ("High Rated Restaurants",
           rotation=30 ,
           fontname="Arial",
           fontweight="bold",
           fontsize=20
          )
plt.ylabel("State", fontname="Arial",
           fontweight="bold",
           fontsize=20,rotation=30)
plt.title("Top 10 States with Rating > 4.5 Restaurants")

plt.gca().invert_xaxis()
plt.show()
```

```python
top10_dishes = (
    df.groupby("Dish_Name", as_index=False)["Price_INR"]
    .max()
    .sort_values("Price_INR", ascending=False)
    .head(10)
)
```

```python
top10_dishes
```

```python
plt.figure(figsize=(12,8))

colors = plt.cm.plasma(np.linspace(0, 1, len(top10_dishes)))

bars = plt.barh(
    top10_dishes["Dish_Name"],
    top10_dishes["Price_INR"],
    color=colors
)

# Value labels at end of bars
for bar in bars:
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height()/2,
        f"₹{int(bar.get_width())}",
        va='center'
    )

plt.xlabel(
    "Price Of Food",
    color="red",
    fontname="Arial",
    fontsize=16,
    fontweight="bold"
)

plt.ylabel(
    "Dish Name",
    color="red",
    fontname="Arial",
    fontsize=16,
    fontweight="bold"
)

plt.title(
    "Top 10 Most Expensive Dishes",
    fontsize=18,
    fontweight="bold"
)

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
top10_mpr = (
    df.groupby("Restaurant_Name")["Rating_Count"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
      .reset_index()
)
top10_mpr
```

```python
plt.figure(figsize=(12, 6))

bars = plt.bar(
    top10_mpr["Restaurant_Name"],
    top10_mpr["Rating_Count"],
    color=plt.cm.viridis(range(len(top10_mpr)))
)

# Add labels (no color here)
plt.bar_label(bars, fmt="%.0f")

plt.title(
    "Top 10 Restaurants by Rating Count",
    fontname="Arial",
    fontsize=20,
    fontweight="bold"
)

plt.xlabel(
    "Restaurant Name",
    fontname="Arial",
    fontsize=20,
    fontweight="bold"
)

plt.ylabel(
    "Rating Count",
    fontname="Arial",
    fontsize=20,
    fontweight="bold"
)

plt.xticks(rotation=20, ha='right')

plt.show()
```
